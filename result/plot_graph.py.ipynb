```python
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

dense_graph_0_1 = np.loadtxt('density_7_sparse_graph.txt').reshape(20, 4)

print(dense_graph_0_1)


plt.plot(np.log10(dense_graph_0_1[:,0]), dense_graph_0_1[:,1])

plt.plot(np.log10(dense_graph_0_1[:,0]), dense_graph_0_1[:,2])

plt.plot(np.log10(dense_graph_0_1[:,0]), dense_graph_0_1[:,3])

plt.title("Le temp d'execution pour le graphe creuse de densite 7")
plt.xlabel("log10(nb vertex du graphe)")
plt.ylabel("temp d'execution")
plt.legend(["Dijkstra naif avec Matrice d'adjacente", "Dijkstra min heap avec Matrice d'adjacente", "Dijkstra min heap avec Liste d'adjacente"])
plt.savefig("Dijkstra resultat avec sparse graphe de densite3.png", transparent = False)
```

```python
plt.plot(dense_graph_0_1[:,0] * dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0], dense_graph_0_1[:,1], 'ro')
m, b = np.polyfit(dense_graph_0_1[:,0] * dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0], dense_graph_0_1[:,1], 1)
plt.plot(dense_graph_0_1[:,0] * dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0],
        m * (dense_graph_0_1[:,0] * dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0]) + b)
plt.title(r"Courbe du temps en fonction de $|V|^2 + |E|$")
plt.xlabel(r'$|V|^2 + |E|$')
#plt.xlabel(r'\textbf{time} (s)')
plt.ylabel("Temp d'execution pour dijkstra naif en matrice d'adjacente")
plt.legend(["graphe creuse avec densite = 7", "fonction regression"])
```

```python
plt.plot((dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0]) * np.log2(dense_graph_0_1[:,0]), dense_graph_0_1[:,3], 'ro')
m, b = np.polyfit((dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0]) * np.log2(dense_graph_0_1[:,0]), dense_graph_0_1[:,1], 1)
plt.plot((dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0]) * np.log2(dense_graph_0_1[:,0]),
        m * (dense_graph_0_1[:,0] + 7 * dense_graph_0_1[:,0]) * np.log2(dense_graph_0_1[:,0]) + b)
plt.title(r"Courbe du temps en fonction de $(|V| + |E|)log(|V|)$")
plt.xlabel(r'$(|V| + |E|)log(|V|)$')
#plt.xlabel(r'\textbf{time} (s)')
plt.ylabel("Temp d'execution pour dijkstra naif en matrice d'adjacente")
plt.legend(["graphe creuse avec densite = 7", "fonction regression"])
```
